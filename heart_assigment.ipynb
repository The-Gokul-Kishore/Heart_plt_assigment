```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polar_plotter():
    theta = np.linspace(0,2*np.pi,100)
    r = 1-np.sin(theta)
    plt.figure(figsize=(10,10))
    plt.polar(theta,r,color='blue',linewidth=2)
    plt.show()
polar_plotter()
```

```python
def parametric_equation_plotter():    
    #defining the values for x and y
    t = np.linspace(0, 2 * np.pi, 1000)
    x = 16 * np.sin(t)**3
    y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
    
    plt.figure(figsize=(8,6))
    plt.plot(x, y, color='red', linewidth=2)
    plt.title("Parametric Heart")
    plt.gca().set_aspect('equal')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
parametric_equation_plotter()
```
